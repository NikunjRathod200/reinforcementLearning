```python
# Define the Deep Q Network architecture
def create_phishing_url_dqn(input_dim, hidden_dim, output_dim):
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(input_dim,)),
        tf.keras.layers.Dense(hidden_dim, activation='relu'),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(output_dim)
    ])

    # Compile the model
    model.compile(optimizer='adam',
                  loss='mean_squared_error')

    return model
```

```python
# Define the Q-learning agent
class QLearningAgent:
    def __init__(self, input_dim, action_dim, hidden_dim=64, learning_rate=0.001, gamma=0.99, epsilon=1.0, epsilon_decay=0.995, epsilon_min=0.01):
        self.input_dim = input_dim
        self.action_dim = action_dim
        self.learning_rate = learning_rate
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.model = create_phishing_url_dqn(input_dim, hidden_dim, action_dim)

    def select_action(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_dim)
        q_values = self.model.predict(np.array(state).reshape(1, -1))
        return np.argmax(q_values[0])

    def train(self, state, action, next_state, reward, done):
        target = reward
        if not done:
            q_values_next = self.model.predict(np.array(next_state).reshape(1, -1))
            target = reward + self.gamma * np.amax(q_values_next[0])
        target_q_values = self.model.predict(np.array(state).reshape(1, -1))
        target_q_values[0][action] = target
        self.model.fit(np.array(state).reshape(1, -1), target_q_values, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# Assuming you have 12 input features and 1 output (0 for legitimate, 1 for phishing)
input_dim = 100
output_dim = 2  # (0 for legitimate, 1 for phishing)
hidden_dim = 128

# Create the Q-learning agent
agent = QLearningAgent(input_dim, output_dim, hidden_dim)

# Training parameters
batch_size = 32
epochs = 10
num_episodes = 100

# Dummy data (replace this with your actual data)
# X_train = np.random.rand(1000, input_dim)
# y_train = np.random.randint(2, size=1000)
```
